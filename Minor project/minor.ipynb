```python
#calculating percentage of all students according to total score
data.head()
data['percentage']=round((data['total score']/300)*100,2)
data.head()
```

```python
data['status'] = data.apply(lambda x : 'fail' if x['maths_status'] == 'Fail' or 
                           x['reading_status'] == 'fail' or x['writing_status'] == 'fail'
                           else 'pass', axis = 1)

data['status'].value_counts().plot.bar(color = 'gray', figsize = (3, 3))
plt.title('overall results')
plt.show()

data.status.value_counts()
#deciding status of students as pass or fail
#if student fails in any one of the subject he is considered as failed candidate else pass.
```

## observation
out of 1000 students 865 students passed in all subjects and 135 students failed in either of the subjects.

```python
# a new coloumn is created with name grades and grades are assigned according to percentage scored
def graded(per):
    if (per>90):
        return 'A'
    elif (per>80 and per<91):
        return 'B'
    elif (per>70 and per<81):
        return 'C'
    elif (per>60 and per<71):
        return 'D'
    elif (per>50 and per<61):
        return 'E'
    else:
        return 'F'
    
    

data['grades']=data.apply(lambda x:graded(x['percentage']),axis=1)
data.head()
```

```python
data['grades'].value_counts().plot.bar(color="darkgreen")
plt.show()
#visualising grades scored by students
```

## Observation 
by analysing above graph, we can conclude that:
   -  A grade, that is top score is achieved by approx 40 students
   - max students scored D grade
   - approximately 110-120 members scored F Grade i.e failed
